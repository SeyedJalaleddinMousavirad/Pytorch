**Vanishing Gradient Problem:**

it happens in Sigmoid and hyperbolic tangant activation functions

The further we go in neural network, the smaller the gradient becomes 

**Solution:** ReLU activation function: max(0,z)

There is a problem in ReLU function, called "dead neuron" problem, which means that gradient in the left half is already vanished!

There are two general approches for dead neuron:


1) in deep learning, this activation function works and the right side, "not vanishging" appears to be enough.

2) definition of modified ReLU function: for example Leaky ReLU 

Also, there is an activation function, called "exponential linear unit" or ELU that can speed up the learning process and leads to a higher accuracy
